```python
#import libraries
import pandas as pd
import numpy as np
from apyori import apriori, load_transactions
```

```python
#load dataset
df = pd.read_csv('dataset/dataset_apriori.csv',header = None)
```

```python
#quicklook
df
```

```python
#converting pandas to list of list
data = []
for i in range(0,len(df)):
    data.append([str(df.values[i,j]) for j in range(0,len(df.columns))])
```

```python
#quick look
data
```

```python
#building apyori model
assosication_rules = apriori(data,min_support=0.5,min_confidence=0.5,min_lift=1.2,min_length=2)
assosication_result = list(assosication_rules)
```

```python
#number of rules
print(len(assosication_result))
```

```python
#result of the rule 
print(assosication_result)
```
