# Collaborative Filtering

This dataset contains user who rating the movies. We want to recommend the movie for user '220' using Collaborative Filtering by KNN and SVD Algorithm

```python
from surprise import Dataset
from surprise import Reader
import pandas as pd
```

```python
# load dataset
df = pd.read_csv('dataset/movie/ratings.csv')
df_movie = pd.read_csv('dataset/movie/movies.csv')
reader = Reader(rating_scale=(1, 5))
```

```python
df.drop('timestamp',axis=1)
```

```python
#load from dataframe
data = Dataset.load_from_df(df[["userId", "movieId", "rating"]], reader)
```

## Algorithms Based on K-Nearest Neighbours (k-NN)

```python
from surprise import KNNWithMeans
```

```python
# # gridsearch
# from surprise.model_selection import GridSearchCV


# sim_options = {
#     "name": ["msd", "cosine"],
#     "min_support": [3, 4, 5],
#     "user_based": [False, True],
# }

# param_grid = {"sim_options": sim_options}

# gs = GridSearchCV(KNNWithMeans, param_grid, measures=["rmse", "mae"], cv=3)
# gs.fit(data)

# print(gs.best_score["rmse"])
# print(gs.best_params["rmse"])
```

```python
algo = KNNWithMeans(sim_options={"name":"msd","user_base":True},k=20)
```

```python
# algo.fit(data)
data_train = data.build_full_trainset()
```

```python
algo.fit(data_train)
```

```python
pred = pd.DataFrame()
# pred['movie'] = 
# pred
```

```python
for i in range(1,len(df_movie)):
    pred.set_value(i,'movieId',i)
    pred.set_value(i,'predRating',algo.predict(220,i).est)
    
    
```

```python
pred.sort_values('predRating',ascending=False)
```

```python
final_pred = pd.merge(left=pred,right=df_movie, left_on='movieId', right_on='movieId')
final_pred.sort_values('predRating',ascending=False).head(10)
```

```python
final_pred.predRating.describe()
```

## Algorithm SVD

```python
from surprise import SVD
# from surprise import Dataset
from surprise.model_selection import GridSearchCV

data = Dataset.load_from_df(df[["userId", "movieId", "rating"]], reader)

param_grid = {
    "n_epochs": [5, 10],
    "lr_all": [0.002, 0.005],
    "reg_all": [0.4, 0.6]
}
gs = GridSearchCV(SVD, param_grid, measures=["rmse", "mae"], cv=3)

gs.fit(data)

print(gs.best_score["rmse"])
print(gs.best_params["rmse"])
```

```python
algoSVD = SVD(n_epochs=10,lr_all=0.005,reg_all=0.4,)
```

```python
algoSVD.fit(data_train)
```

```python
predSVD = pd.DataFrame()
```

```python
for i in range(1,len(df_movie)):
    predSVD.set_value(i,'movieId',i)
    predSVD.set_value(i,'predRating',algoSVD.predict(220,i).est)

```

```python
predSVD.sort_values('predRating',ascending=False)
```

```python
final_predSVD = pd.merge(left=predSVD,right=df_movie, left_on='movieId', right_on='movieId')
final_predSVD.sort_values('predRating',ascending=False).head(10)
```
